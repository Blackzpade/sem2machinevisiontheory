```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load image
image_path = r'C:\Users\acer\Desktop\a2.jpg'
img = Image.open(image_path)

# Translation: Shift image by (x, y) pixels
def translate_image(image, x_offset, y_offset):
    translation_matrix = np.array([[1, 0, x_offset],
                                   [0, 1, y_offset]])
    return image.transform(image.size, Image.AFFINE, data=translation_matrix.flatten(), resample=Image.BICUBIC)

# Rotation: Rotate image by angle in degrees
def rotate_image(image, angle):
    return image.rotate(angle)

# Flipping: Flip image horizontally or vertically
def flip_image(image, direction='horizontal'):
    if direction == 'horizontal':
        return image.transpose(Image.FLIP_LEFT_RIGHT)
    elif direction == 'vertical':
        return image.transpose(Image.FLIP_TOP_BOTTOM)
    else:
        raise ValueError("Direction must be 'horizontal' or 'vertical'")

# Function to display images in the notebook
def display_image(image, title="Image"):
    plt.imshow(image)
    plt.axis('off')  # Hide axes
    plt.title(title)
    plt.show()

# Example usage

# Translate the image by 100 pixels in both x and y directions
translated_img = translate_image(img, 100, 100)
display_image(translated_img, title="Translated Image")

# Rotate the image by 45 degrees
rotated_img = rotate_image(img, 45)
display_image(rotated_img, title="Rotated Image")

# Flip the image horizontally
flipped_img = flip_image(img, 'horizontal')
display_image(flipped_img, title="Flipped Image (Horizontal)")

# Flip the image vertically
flipped_img_vertical = flip_image(img, 'vertical')
display_image(flipped_img_vertical, title="Flipped Image (Vertical)")

# Save the transformed images (optional)
translated_img.save('translated_image.jpg')
rotated_img.save('rotated_image.jpg')
flipped_img.save('flipped_image_horizontal.jpg')
flipped_img_vertical.save('flipped_image_vertical.jpg')
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load image
image_path = r'C:\Users\acer\Desktop\a2.jpg'  # Update your image path here
img = Image.open(image_path)

# Resize the image (specify width and height)
def resize_image(image, width, height):
    return image.resize((width, height))

# Scale the image (by a scaling factor, e.g., 0.5 for 50% of the original size)
def scale_image(image, scale_factor):
    width, height = image.size
    new_size = (int(width * scale_factor), int(height * scale_factor))
    return image.resize(new_size)

# Crop the image (specify the left, upper, right, and lower pixel coordinates)
def crop_image(image, left, upper, right, lower):
    return image.crop((left, upper, right, lower))

# Function to display images in the notebook
def display_image(image, title="Image"):
    plt.imshow(image)
    plt.axis('off')  # Hide axes
    plt.title(title)
    plt.show()

# Example usage:

# 1. Resize image to 400x400 pixels
resized_img = resize_image(img, 400, 400)
display_image(resized_img, title="Resized Image")

# 2. Scale image by 50% (0.5 scaling factor)
scaled_img = scale_image(img, 0.5)
display_image(scaled_img, title="Scaled Image")

# 3. Crop image to a region: left=100, upper=100, right=400, lower=400
cropped_img = crop_image(img, 100, 100, 400, 400)
display_image(cropped_img, title="Cropped Image")

# Save the transformed images (optional)
resized_img.save('resized_image.jpg')
scaled_img.save('scaled_image.jpg')
cropped_img.save('cropped_image.jpg')

```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the grayscale image
image_path = r'C:\Users\acer\Desktop\a2.jpg'  # Replace with your image path
image = Image.open(image_path).convert('L')  # Convert image to grayscale

# Convert image to numpy array
gray_image = np.array(image)

# Set a threshold (this can be changed depending on your image)
threshold = 128  # Range is 0-255

# Create a binary image based on the threshold
binary_image = np.where(gray_image >= threshold, 255, 0)

# Display the binary image
plt.imshow(binary_image, cmap='gray')
plt.axis('off')  # Turn off the axis
plt.show()

# Optionally save the binary image
output_image = Image.fromarray(binary_image.astype(np.uint8))
output_image.save('binary_image.png')
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def find_complement(image_path):
    # Load the image
    image = Image.open(image_path)
    
    # Check if the image is grayscale or color (RGB)
    if image.mode == 'L':  # Grayscale image
        print("Processing grayscale image...")
        
        # Convert grayscale image to numpy array
        image_array = np.array(image)
        
        # Find the complement (invert pixel values)
        complement_image = 255 - image_array
        
        # Display the complement of the grayscale image
        plt.imshow(complement_image, cmap='gray')
        plt.axis('off')  # Turn off the axis
        plt.title("Complement of Grayscale Image")
        plt.show()
        
        # Save the complement image
        output_image = Image.fromarray(complement_image.astype(np.uint8))
        output_image.save('complement_grayscale_image.png')
    
    elif image.mode == 'RGB':  # Color image
        print("Processing color image...")
        
        # Convert color image to numpy array (RGB)
        image_array = np.array(image)
        
        # Find the complement (invert pixel values for each channel)
        complement_image = 255 - image_array
        
        # Display the complement of the color image
        plt.imshow(complement_image)
        plt.axis('off')  # Turn off the axis
        plt.title("Complement of Color Image")
        plt.show()
        
        # Save the complement image
        output_image = Image.fromarray(complement_image.astype(np.uint8))
        output_image.save('complement_color_image.png')
    
    else:
        print("Unsupported image format. Only grayscale and RGB are supported.")

# Example usage
image_path = r'C:\Users\acer\Desktop\a2.jpg'  # Replace with the path to your image
find_complement(image_path)

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show_image(image, title="Image"):
    """Utility function to display an image"""
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

def manipulate_color_spaces(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # 1. RGB to HSV conversion
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    show_image(image_hsv, "HSV Image")

    # 2. HSV back to RGB conversion
    image_rgb_from_hsv = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2BGR)
    show_image(image_rgb_from_hsv, "RGB from HSV")

    # 3. RGB to LAB conversion
    image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    show_image(image_lab, "LAB Image")

    # 4. LAB back to RGB conversion
    image_rgb_from_lab = cv2.cvtColor(image_lab, cv2.COLOR_LAB2BGR)
    show_image(image_rgb_from_lab, "RGB from LAB")

# Example usage
image_path = r'C:\Users\acer\Desktop\a2.jpg'  # Replace with the path to your image
manipulate_color_spaces(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show_image(image, title="Image"):
    """Utility function to display an image"""
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

def split_and_merge_channels(image_path):
    # Load the image (OpenCV loads in BGR format)
    image = cv2.imread(image_path)

    # Split the image into its BGR channels
    (blue, green, red) = cv2.split(image)
    
    # Display each individual channel
    show_image(blue, "Blue Channel")
    show_image(green, "Green Channel")
    show_image(red, "Red Channel")

    # Merge the channels back together (in BGR order)
    merged_image = cv2.merge([blue, green, red])

    # Display the merged image
    show_image(merged_image, "Merged Image (BGR)")

    # Example: Creating a modified image by manipulating channels
    # In this case, we will just swap the Red and Blue channels
    swapped_image = cv2.merge([red, green, blue])

    # Display the swapped image (RGB will look different after swap)
    show_image(swapped_image, "Swapped Red and Blue Channels")

# Example usage
image_path = r'C:\Users\acer\Desktop\a2.jpg'  # Replace with the path to your image
split_and_merge_channels(image_path)
```
